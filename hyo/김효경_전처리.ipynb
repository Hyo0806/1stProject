# 행정동 데이터

```python
code_all = pd.read_excel('data/행정동코드.xlsx')
code_sw_pd = code_all[code_all['시군구명']=='수원시 팔달구'].reset_index(drop=True)
code_sw_pd = code_sw_pd.drop(0).reset_index(drop=True)
code_sw_pd['행정동코드'] = code_sw_pd['행정동코드']//100
code_sw_pd
```

# 수원 팔달구 데이터

```python
sw_pd = df_23_25[df_23_25['admi_cty_no'].isin(code_sw_pd_num)]
sw_pd = sw_pd.reset_index(drop=True)
sw_pd
```

```python
not_use = ['cty_rgn_no', 'card_tpbuz_cd', 'card_tpbuz_nm_1', 'card_tpbuz_nm_2']
sw_pd.drop(not_use, axis=1, inplace=True)
sw_pd
```

```python
sw_pd.columns = ['기준년월일', '행정동코드', '시간대', '성별', '연령별', '요일', '매출금액', '매출건수']
sw_pd.head(10)
```

```python
sw_pd_gender_sum = (sw_pd.groupby(['기준년월일', '행정동코드', '시간대', '연령별', '요일'], as_index=False)[['매출금액','매출건수']].sum())
sw_pd_gender_sum.head(10)
```
