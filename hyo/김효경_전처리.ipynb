```python
sw_rainfall = pd.concat([sw_rainfall_1, sw_rainfall_2, sw_rainfall_3], ignore_index=True)
sw_rainfall = sw_rainfall[sw_rainfall['rainfall'] != 0].reset_index(drop=True)
sw_rainfall
```

```python
h = (sw_rainfall['hour'] // 100).astype(int)

def hour_to_slot(x):
    if   0  <= x <= 6: return 1    # 00:00~06:59
    elif 7  <= x <= 8: return 2    # 07:00~08:59
    elif 9  <= x <= 10:return 3    # 09:00~10:59
    elif 11 <= x <= 12:return 4    # 11:00~12:59
    elif 13 <= x <= 14:return 5    # 13:00~14:59
    elif 15 <= x <= 16:return 6    # 15:00~16:59
    elif 17 <= x <= 18:return 7    # 17:00~18:59
    elif 19 <= x <= 20:return 8    # 19:00~20:59
    elif 21 <= x <= 22:return 9    # 21:00~22:59
    elif x == 23:      return 10   # 23:00~23:59
    else:              return None   

sw_rainfall['time_slot'] = h.map(hour_to_slot)
sw_rainfall.columns = ['시간', '강수량', '기준년월일', '시간대']
sw_rainfall = sw_rainfall[['기준년월일', '시간대', '시간', '강수량']]
sw_rainfall
```

```python
sw_rainfall_group = (sw_rainfall.groupby(['기준년월일', '시간대'], as_index=False)['강수량'].mean())
sw_rainfall_group = sw_rainfall_group[sw_rainfall_group['강수량'] >= 0].reset_index(drop=True)
sw_rainfall_group
```
