```python
%%time

import os
import glob
import re
import pandas as pd

# =========================
# 동 리스트(44개)
# =========================
DONG_LIST = [
    "파장동","율천동","정자1동","정자2동","정자3동","영화동","송죽동","조원1동","조원2동","연무동",
    "세류1동","세류2동","세류3동","평동","서둔동","구운동","금곡동","호매실동","권선1동","권선2동",
    "곡선동","입북동","매교동","매산동","고등동","화서1동","화서2동","지동","우만1동","우만2동",
    "인계동","행궁동","매탄1동","매탄2동","매탄3동","매탄4동","원천동","영통1동","영통2동",
    "영통3동","망포1동","망포2동","광교1동","광교2동",
]

BASE_DIR = "data/수원시 동 기상"

# =========================
# rain 컬럼 탐색
# =========================
def detect_rain_col(df: pd.DataFrame) -> str:
    cands = [c for c in df.columns if str(c).strip().lower().startswith("value location")]
    if cands:
        return cands[0]
    if "rain" in df.columns:
        return "rain"
    raise ValueError("rain 컬럼을 찾지 못함")

# =========================
# 연월 → 말일
# =========================
def last_day_of_month(ym: str) -> str:
    dt = pd.to_datetime(ym + "01", format="%Y%m%d")
    return (dt + pd.offsets.MonthEnd(1)).strftime("%Y%m%d")

# =========================
# hour 롤오버 기반 date 생성 (강수 전용)
# =========================
def add_date_by_hour_rollover(df: pd.DataFrame, start_ymd: str, end_ymd: str):
    df = df.copy()

    start_dt = pd.to_datetime(start_ymd, format="%Y%m%d")
    end_dt   = pd.to_datetime(end_ymd, format="%Y%m%d")

    # 🔥 hour 완전 정규화 (문자열/float/NaN 전부 대응)
    df["hour"] = (
        pd.to_numeric(df["hour"], errors="coerce")
        .fillna(0)
        .astype(int)
    )

    dates = []
    cur_date = start_dt
    prev_hour = None

    for h in df["hour"]:
        if prev_hour is not None and h < prev_hour:
            cur_date += pd.Timedelta(days=1)
        dates.append(cur_date)
        prev_hour = h

    df["date"] = pd.to_datetime(dates)

    # 종료일 컷
    df = df[df["date"] <= end_dt]

    df["date"] = df["date"].dt.strftime("%Y%m%d")
    return df.reset_index(drop=True)

# =========================
# 동 하나의 강수 데이터 로드
# =========================
def load_one_dong_rain(dong: str) -> pd.DataFrame:
    pattern = os.path.join(BASE_DIR, f"{dong}_강수_*.csv")
    files = sorted(glob.glob(pattern))

    if not files:
        print(f"[WARN] 파일 없음: {dong}")
        return pd.DataFrame()

    parts = []

    for fp in files:
        df = pd.read_csv(fp)

        # Start 행 제거
        if " format: day" in df.columns:
            df = df[~df[" format: day"].astype(str).str.contains("Start", na=False)]
            df = df.drop(columns=[" format: day"], errors="ignore")

        # rain 컬럼 통일
        rain_col = detect_rain_col(df)
        df = df.rename(columns={rain_col: "rain"})

        # 파일명 기간 파싱
        m = re.search(r"_(\d{6})_(\d{6})\.csv$", os.path.basename(fp))
        if not m:
            raise ValueError(f"파일명 파싱 실패: {fp}")

        start_ymd = m.group(1) + "01"
        end_ymd   = last_day_of_month(m.group(2))

        # date 생성
        df = add_date_by_hour_rollover(df, start_ymd, end_ymd)

        df["DONG"] = dong
        parts.append(df)

    out = pd.concat(parts, ignore_index=True)
    return out[["DONG", "date", "hour", "rain"]]

# =========================
# 전체 동 로드
# =========================
all_dfs = []
for dong in DONG_LIST:
    df_d = load_one_dong_rain(dong)
    if not df_d.empty:
        all_dfs.append(df_d)

sw_rain = pd.concat(all_dfs, ignore_index=True)

print("[DONE] sw_rain rows:", len(sw_rain))
print(sw_rain.head())

# =========================
# 시간대 슬롯 생성
# =========================
sw_rain["hour_int"] = (sw_rain["hour"] // 100).astype(int)
sw_rain["HOUR"] = sw_rain["hour_int"].apply(hour_to_slot)
```

```python
import pandas as pd

KEYS = ["TA_YMD", "DONG", "HOUR"]

# 0) 타입/문자 정리 (필수: 키 타입 불일치로 매칭 안 되는 것 방지)
sw_koreanfood["TA_YMD"] = sw_koreanfood["TA_YMD"].astype("int64")
sw_koreanfood["HOUR"]   = sw_koreanfood["HOUR"].astype("int64")
sw_koreanfood["DONG"]   = sw_koreanfood["DONG"].astype(str).str.strip()

sw_temp["TA_YMD"] = sw_temp["TA_YMD"].astype("int64")
sw_temp["HOUR"]   = sw_temp["HOUR"].astype("int64")
sw_temp["DONG"]   = sw_temp["DONG"].astype(str).str.strip()

sw_rain["TA_YMD"] = sw_rain["TA_YMD"].astype("int64")
sw_rain["HOUR"]   = sw_rain["HOUR"].astype("int64")
sw_rain["DONG"]   = sw_rain["DONG"].astype(str).str.strip()

# 1) temp/rain 키 중복 제거(집계) -> 1키=1행 만들기
#    TEMP: 같은 슬롯 내 평균(원하면 max/min 등으로 바꿔도 됨)
temp_agg = (sw_temp
            .groupby(KEYS, as_index=False)["TEMP"]
            .mean())

#    RAIN: 같은 슬롯 내 평균 (원하면 sum/max 등으로 바꿔도 됨)
rain_agg = (sw_rain
            .groupby(KEYS, as_index=False)["RAIN"]
            .mean())

# 2) sw_koreanfood 기준 LEFT MERGE (행 수 유지)
n0 = len(sw_koreanfood)

out = sw_koreanfood.merge(
    temp_agg, on=KEYS, how="left", validate="m:1"
).merge(
    rain_agg, on=KEYS, how="left", validate="m:1"
)

# 3) 행 수 검증(무조건 동일해야 함)
assert len(out) == n0, f"행 수 증가/감소 발생! before={n0}, after={len(out)}"

# 결과
sw_koreanfood = out
print("OK. rows 동일:", len(sw_koreanfood))
print(sw_koreanfood[["TA_YMD","DONG","HOUR","TEMP","RAIN"]].head())
```
