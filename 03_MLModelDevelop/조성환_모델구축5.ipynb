```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
import warnings
from datetime import datetime, timedelta
import json

warnings.filterwarnings('ignore')

# ë„¤íŠ¸ì›Œí¬ê°€ ì œí•œëœ í™˜ê²½ì—ì„œ ì‹œë®¬ë ˆì´ì…˜ì„ ìœ„í•œ ë”ë¯¸ í•¨ìˆ˜
def get_weather_data_dummy(date_str, location="ìˆ˜ì›"):
    """ê¸°ìƒ ë°ì´í„° ì‹œë®¬ë ˆì´ì…˜ (ì‹¤ì œ í™˜ê²½ì—ì„œëŠ” API í˜¸ì¶œ)"""
    date = datetime.strptime(date_str, '%Y-%m-%d')
    month = date.month
    
    # ì›”ë³„ í‰ê·  ì˜¨ë„ íŒ¨í„´ (ì‹¤ì œë¡œëŠ” APIì—ì„œ ê°€ì ¸ì˜´)
    temp_pattern = {
        1: -1.0, 2: 0.5, 3: 6.5, 4: 12.5, 5: 16.8, 6: 22.0,
        7: 26.1, 8: 26.5, 9: 22.8, 10: 14.8, 11: 6.5, 12: -1.0
    }
    
    # ì›”ë³„ í‰ê·  ê°•ìˆ˜ëŸ‰ íŒ¨í„´
    rain_pattern = {
        1: 0.04, 2: 0.04, 3: 0.04, 4: 0.09, 5: 0.15, 6: 0.15,
        7: 0.59, 8: 0.35, 9: 0.42, 10: 0.15, 11: 0.18, 12: 0.06
    }
    
    # ì•½ê°„ì˜ ëœë¤ ë³€ë™ ì¶”ê°€
    temp = temp_pattern.get(month, 15.0) + np.random.uniform(-2, 2)
    rain = rain_pattern.get(month, 0.1) + max(0, np.random.uniform(-0.05, 0.1))
    
    return {
        'temperature': round(temp, 1),
        'rainfall': round(rain, 2),
        'source': 'simulated'
    }


class WeatherAPIPredictionSystem:
    """ê¸°ìƒì²­ API ì—°ë™ ì˜ˆì¸¡ ì‹œìŠ¤í…œ"""
    
    def __init__(self, data_path, api_key=None):
        """
        Args:
            data_path: í•™ìŠµ ë°ì´í„° CSV ê²½ë¡œ

            api_key: ê¸°ìƒì²­ API í‚¤ (ì„ íƒì‚¬í•­)
        """
        print("=" * 80)
        print("ê¸°ìƒì²­ API ì—°ë™ ì˜ˆì¸¡ ì‹œìŠ¤í…œ ì´ˆê¸°í™” ì¤‘...")
        print("=" * 80)
        load_dotenv()
        self.api_key = os.getenv('API_KEY')
        self.df = pd.read_csv('C:/taro/1ì°¨_í”„ë¡œì íŠ¸/í†µí•©ë°ì´í„°/SUWON_S_DATA_TABLE.csv')
        self.scaler_X = StandardScaler()
        
        # ë°ì´í„° ì „ì²˜ë¦¬
        self._preprocess_data()
        
        # ëª¨ë¸ í•™ìŠµ
        self._train_models()
        
        print("\nâœ“ ì‹œìŠ¤í…œ ì¤€ë¹„ ì™„ë£Œ!")
        if api_key:
            print(f"âœ“ ê¸°ìƒì²­ API í‚¤ ì„¤ì •ë¨")
        else:
            print("âš  API í‚¤ ë¯¸ì„¤ì • - ì‹œë®¬ë ˆì´ì…˜ ëª¨ë“œë¡œ ì‘ë™")
        print("=" * 80)
    
    def _preprocess_data(self):
        """ë°ì´í„° ì „ì²˜ë¦¬"""
        self.df['TA_YMD'] = pd.to_datetime(self.df['TA_YMD'], format='%Y%m%d')
        self.df['MONTH'] = self.df['TA_YMD'].dt.month
        self.df['DAY_OF_YEAR'] = self.df['TA_YMD'].dt.dayofyear
        
        # ì¼ë³„ ì§‘ê³„
        self.daily_stats = self.df.groupby(['TA_YMD', 'DAY', 'TEMP', 'RAIN']).agg({
            'UNIT': 'sum',
            'CNT': 'sum',
            'MONTH': 'first',
            'DAY_OF_YEAR': 'first'
        }).reset_index()
        
        print(f"\ní•™ìŠµ ë°ì´í„°: {len(self.daily_stats):,}ì¼")
    
    def _train_models(self):
        """ëª¨ë¸ í•™ìŠµ"""
        features = ['DAY', 'TEMP', 'RAIN', 'MONTH', 'DAY_OF_YEAR']
        X = self.daily_stats[features]
        y_unit = self.daily_stats['UNIT']
        y_cnt = self.daily_stats['CNT']
        
        X_train, X_test, y_unit_train, y_unit_test, y_cnt_train, y_cnt_test = train_test_split(
            X, y_unit, y_cnt, test_size=0.2, random_state=42
        )
        
        X_train_scaled = self.scaler_X.fit_transform(X_train)
        
        # ë§¤ì¶œì•¡ ëª¨ë¸
        self.model_unit = GradientBoostingRegressor(
            n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42
        )
        self.model_unit.fit(X_train_scaled, y_unit_train)
        
        # ì´ìš©ê³ ê°ìˆ˜ ëª¨ë¸
        self.model_cnt = GradientBoostingRegressor(
            n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42
        )
        self.model_cnt.fit(X_train_scaled, y_cnt_train)
        
        print("âœ“ ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ í•™ìŠµ ì™„ë£Œ")
    
    def get_kma_weather(self, date_str, location="ìˆ˜ì›"):
        """
        ê¸°ìƒì²­ APIë¥¼ í†µí•´ ì‹¤ì‹œê°„ ë‚ ì”¨ ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
        
        ì‹¤ì œ ì‚¬ìš© ì‹œ í•„ìš”í•œ API:
        - ê¸°ìƒì²­ ë‹¨ê¸°ì˜ˆë³´ API: https://data.go.kr/
        - OpenWeatherMap API: https://openweathermap.org/api
        
        Args:
            date_str: ë‚ ì§œ (YYYY-MM-DD)
            location: ì§€ì—­ëª… (ê¸°ë³¸: ìˆ˜ì›)
        
        Returns:
            dict: {'temperature': float, 'rainfall': float, 'source': str}
        """
        
        # ì‹¤ì œ API í˜¸ì¶œ ì½”ë“œ (ì˜ˆì‹œ)
        if self.api_key:
            try:
                # ê¸°ìƒì²­ API í˜¸ì¶œ ì˜ˆì‹œ ì½”ë“œ
                # import requests
                # 
                # # ê¸°ìƒì²­ ë‹¨ê¸°ì˜ˆë³´ API ì—”ë“œí¬ì¸íŠ¸
                # url = "http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getVilageFcst"
                # 
                # # ìˆ˜ì›ì‹œ ê²©ì ì¢Œí‘œ (nx=60, ny=121)
                # params = {
                #     'serviceKey': self.api_key,
                #     'pageNo': '1',
                #     'numOfRows': '1000',
                #     'dataType': 'JSON',
                #     'base_date': date_str.replace('-', ''),
                #     'base_time': '0500',
                #     'nx': '60',
                #     'ny': '121'
                # }
                # 
                # response = requests.get(url, params=params)
                # data = response.json()
                # 
                # # ì‘ë‹µì—ì„œ ì˜¨ë„(TMP)ì™€ ê°•ìˆ˜ëŸ‰(RN1) ì¶”ì¶œ
                # temp = None
                # rain = 0
                # 
                # for item in data['response']['body']['items']['item']:
                #     if item['category'] == 'TMP':
                #         temp = float(item['fcstValue'])
                #     elif item['category'] == 'RN1':
                #         rain_str = item['fcstValue']
                #         rain = 0 if rain_str == 'ê°•ìˆ˜ì—†ìŒ' else float(rain_str.replace('mm', ''))
                # 
                # return {
                #     'temperature': temp,
                #     'rainfall': rain,
                #     'source': 'KMA API'
                # }
                
                # ë„¤íŠ¸ì›Œí¬ ì œí•œìœ¼ë¡œ ì‹œë®¬ë ˆì´ì…˜ ì‚¬ìš©
                print("âš  ë„¤íŠ¸ì›Œí¬ ì œí•œìœ¼ë¡œ ì‹œë®¬ë ˆì´ì…˜ ëª¨ë“œ ì‚¬ìš©")
                return get_weather_data_dummy(date_str, location)
                
            except Exception as e:
                print(f"âš  API í˜¸ì¶œ ì˜¤ë¥˜: {e}")
                print("ì‹œë®¬ë ˆì´ì…˜ ë°ì´í„° ì‚¬ìš©")
                return get_weather_data_dummy(date_str, location)
        else:
            # API í‚¤ê°€ ì—†ìœ¼ë©´ ì‹œë®¬ë ˆì´ì…˜ ë°ì´í„° ì‚¬ìš©
            return get_weather_data_dummy(date_str, location)
    
    def predict_with_api(self, date_str, use_api=True, manual_temp=None, manual_rain=None):
        """
        API ì—°ë™ ì˜ˆì¸¡ ìˆ˜í–‰
        
        Args:
            date_str: ì˜ˆì¸¡ ë‚ ì§œ (YYYY-MM-DD)
            use_api: API ì‚¬ìš© ì—¬ë¶€ (Falseë©´ ìˆ˜ë™ ì…ë ¥ ì‚¬ìš©)
            manual_temp: ìˆ˜ë™ ì…ë ¥ ì˜¨ë„
            manual_rain: ìˆ˜ë™ ì…ë ¥ ê°•ìˆ˜ëŸ‰
        """
        print("\n" + "=" * 80)
        print("ì˜ˆì¸¡ ì‹œì‘...")
        print("=" * 80)
        
        # ë‚ ì§œ íŒŒì‹±
        date = pd.to_datetime(date_str, format='%Y-%m-%d')
        day_of_week = date.dayofweek + 1
        month = date.month
        day_of_year = date.dayofyear
        day_names = ['ì›”ìš”ì¼', 'í™”ìš”ì¼', 'ìˆ˜ìš”ì¼', 'ëª©ìš”ì¼', 'ê¸ˆìš”ì¼', 'í† ìš”ì¼', 'ì¼ìš”ì¼']
        
        # ë‚ ì”¨ ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
        if use_api and (manual_temp is None or manual_rain is None):
            print("\nğŸŒ¤ï¸  ì‹¤ì‹œê°„ ê¸°ìƒ ë°ì´í„° ê°€ì ¸ì˜¤ëŠ” ì¤‘...")
            weather_data = self.get_kma_weather(date_str)
            temp = weather_data['temperature']
            rain = weather_data['rainfall']
            data_source = weather_data['source']
            print(f"âœ“ ê¸°ìƒ ë°ì´í„° ì†ŒìŠ¤: {data_source}")
        else:
            temp = manual_temp if manual_temp is not None else \
                   self.daily_stats[self.daily_stats['MONTH'] == month]['TEMP'].mean()
            rain = manual_rain if manual_rain is not None else \
                   self.daily_stats[self.daily_stats['MONTH'] == month]['RAIN'].mean()
            data_source = "manual input"
            print(f"âœ“ ìˆ˜ë™ ì…ë ¥ ë°ì´í„° ì‚¬ìš©")
        
        # ì˜ˆì¸¡ ìˆ˜í–‰
        features = np.array([[day_of_week, temp, rain, month, day_of_year]])
        features_scaled = self.scaler_X.transform(features)
        
        pred_unit = self.model_unit.predict(features_scaled)[0]
        pred_cnt = self.model_cnt.predict(features_scaled)[0]
        
        # ê²°ê³¼ ì¶œë ¥
        print(f"\nğŸ“… ì…ë ¥ ë‚ ì§œ: {date_str} ({day_names[day_of_week-1]})")
        print(f"ğŸŒ¡ï¸  ì˜¨ë„: {temp:.1f}Â°C")
        print(f"ğŸŒ§ï¸  ê°•ìˆ˜ëŸ‰: {rain:.2f}mm")
        print(f"ğŸ“Š ë°ì´í„° ì†ŒìŠ¤: {data_source}")
        print("\n" + "-" * 80)
        print("ğŸ¯ ì˜ˆì¸¡ ê²°ê³¼:")
        print("-" * 80)
        print(f"ğŸ’° ì˜ˆìƒ ë§¤ì¶œì•¡(UNIT): {pred_unit:,.0f}ì›")
        print(f"ğŸ‘¥ ì˜ˆìƒ ì´ìš©ê³ ê°ìˆ˜(CNT): {pred_cnt:,.0f}ëª…")
        print("=" * 80)
        
        return {
            'date': date_str,
            'day_of_week': day_names[day_of_week-1],
            'temperature': temp,
            'rainfall': rain,
            'predicted_sales': pred_unit,
            'predicted_customers': pred_cnt,
            'data_source': data_source
        }
    
    def predict_multiple_days(self, start_date, days=7, use_api=True):
        """
        ì—¬ëŸ¬ ë‚ ì§œ ì—°ì† ì˜ˆì¸¡
        
        Args:
            start_date: ì‹œì‘ ë‚ ì§œ (YYYY-MM-DD)
            days: ì˜ˆì¸¡ ì¼ìˆ˜
            use_api: API ì‚¬ìš© ì—¬ë¶€
        """
        print("\n" + "=" * 80)
        print(f"{days}ì¼ ì—°ì† ì˜ˆì¸¡")
        print("=" * 80)
        
        start = datetime.strptime(start_date, '%Y-%m-%d')
        results = []
        
        for i in range(days):
            current_date = start + timedelta(days=i)
            date_str = current_date.strftime('%Y-%m-%d')
            
            result = self.predict_with_api(date_str, use_api=use_api)
            results.append(result)
        
        # ìš”ì•½ í…Œì´ë¸” ì¶œë ¥
        print("\n" + "=" * 80)
        print(f"{days}ì¼ ì˜ˆì¸¡ ìš”ì•½")
        print("=" * 80)
        print(f"{'ë‚ ì§œ':<12} {'ìš”ì¼':<8} {'ì˜¨ë„(Â°C)':<10} {'ê°•ìˆ˜(mm)':<10} {'ì˜ˆìƒë§¤ì¶œ(ì›)':<15} {'ì˜ˆìƒê³ ê°(ëª…)':<12}")
        print("-" * 80)
        
        for r in results:
            print(f"{r['date']:<12} {r['day_of_week']:<8} {r['temperature']:<10.1f} "
                  f"{r['rainfall']:<10.2f} {r['predicted_sales']:<15,.0f} {r['predicted_customers']:<12,.0f}")
        
        print("=" * 80)
        
        return results
    
    def run_interactive(self):
        """ì¸í„°ë™í‹°ë¸Œ ëª¨ë“œ"""
        print("\n" + "=" * 80)
        print("ğŸŒ¤ï¸  ê¸°ìƒì²­ API ì—°ë™ ì˜ˆì¸¡ ì‹œìŠ¤í…œ - ì¸í„°ë™í‹°ë¸Œ ëª¨ë“œ")
        print("=" * 80)
        print("\nì‚¬ìš© ë°©ë²•:")
        print("  1. ë‚ ì§œ ì…ë ¥ (YYYY-MM-DD)")
        print("  2. API ì‚¬ìš© ì—¬ë¶€ ì„ íƒ (y/n)")
        print("  3. ìˆ˜ë™ ì…ë ¥ ì‹œ ì˜¨ë„/ê°•ìˆ˜ëŸ‰ ì…ë ¥")
        print("  4. ì¢…ë£Œ: 'q' ë˜ëŠ” 'quit'")
        print("=" * 80)
        
        while True:
            print("\n" + "-" * 80)
            date_input = input("ğŸ“… ë‚ ì§œ ì…ë ¥ (YYYY-MM-DD): ").strip()
            
            if date_input.lower() in ['q', 'quit', 'ì¢…ë£Œ']:
                print("\nì‹œìŠ¤í…œ ì¢…ë£Œ")
                break
            
            # API ì‚¬ìš© ì—¬ë¶€
            api_choice = input("ğŸŒ¤ï¸  ì‹¤ì‹œê°„ ê¸°ìƒ API ì‚¬ìš©? (y/n, ê¸°ë³¸: y): ").strip().lower()
            use_api = api_choice != 'n'
            
            if not use_api:
                temp_input = input("ğŸŒ¡ï¸  ì˜¨ë„ ì…ë ¥ (Â°C): ").strip()
                rain_input = input("ğŸŒ§ï¸  ê°•ìˆ˜ëŸ‰ ì…ë ¥ (mm): ").strip()
                
                temp = float(temp_input) if temp_input else None
                rain = float(rain_input) if rain_input else None
                
                self.predict_with_api(date_input, use_api=False, 
                                     manual_temp=temp, manual_rain=rain)
            else:
                self.predict_with_api(date_input, use_api=True)


def main():
    """ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜"""
    print("""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘                   ìˆ˜ì›ì‹œ ë§¤ì¶œì•¡ ì˜ˆì¸¡ ì‹œìŠ¤í…œ - ê¸°ìƒì²­ API ì—°ë™                     â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    """)

    api_key = os.getenv('API_KEY')
    
    # ì‹œìŠ¤í…œ ì´ˆê¸°í™”
    system = WeatherAPIPredictionSystem(
        'C:/taro/1ì°¨_í”„ë¡œì íŠ¸/í†µí•©ë°ì´í„°/SUWON_S_DATA_TABLE.csv',
        api_key=api_key
    )
    
    # ì‚¬ìš© ì˜ˆì‹œ
    print("\n" + "=" * 80)
    print("ì‚¬ìš© ì˜ˆì‹œ")
    print("=" * 80)
    
    # ì˜ˆì‹œ 1: ë‹¨ì¼ ë‚ ì§œ ì˜ˆì¸¡ (API ì‚¬ìš©)
    print("\n[ì˜ˆì‹œ 1] ë‹¨ì¼ ë‚ ì§œ ì˜ˆì¸¡ (ì‹¤ì‹œê°„ ê¸°ìƒ ë°ì´í„°)")
    system.predict_with_api('2026-01-15', use_api=True)
    
    # ì˜ˆì‹œ 2: ìˆ˜ë™ ì…ë ¥
    print("\n[ì˜ˆì‹œ 2] ìˆ˜ë™ ì…ë ¥ ì˜ˆì¸¡")
    system.predict_with_api('2026-03-20', use_api=False, 
                           manual_temp=12.5, manual_rain=1.0)
    
    # ì˜ˆì‹œ 3: 7ì¼ ì—°ì† ì˜ˆì¸¡
    print("\n[ì˜ˆì‹œ 3] 7ì¼ ì—°ì† ì˜ˆì¸¡")
    system.predict_multiple_days('2026-02-01', days=7, use_api=True)
    
    # ì¸í„°ë™í‹°ë¸Œ ëª¨ë“œ ì‹¤í–‰ ì—¬ë¶€
    print("\n" + "=" * 80)
    choice = input("ì¸í„°ë™í‹°ë¸Œ ëª¨ë“œë¡œ ì „í™˜í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ").strip().lower()
    if choice == 'y':
        system.run_interactive()
    else:
        print("\ní”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")


if __name__ == "__main__":
    main()
```
